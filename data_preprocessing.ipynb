# open access

```python
    from functools import reduce

#for file in file_list :

    #Import data
    data = pd.read_excel(f'D:\MARBURG VIRUS DISEASES/data/WOSCC/woscc_bon_23_02_verif_on_wos.xls')[['UT (Unique WOS ID)','Open Access Designations']];

    # set accession number as index
    data = data.set_index('UT (Unique WOS ID)')

    #
    df_expand = data['Open Access Designations'].str.split(', ', expand=True)

    ## get dummies to have contries on columns
    dfs = [pd.get_dummies(df_expand[col]) for col in df_expand]
    d = reduce(lambda x,y : x.add(y, fill_value=0), dfs)

    # data cleaning
    cols = [c for c in d.columns if c[-1:] != '.']
    df=d[cols]
    cols = [c for c in df.columns.to_list() if len(c) != 1]
    df=df[cols]

    # concat
    df2 = pd.concat([data, df], axis=1)
    df2.replace(0, np.nan, inplace=True)

    #save data
    df2.to_csv(f'../DATA_PREPROCESSING/open_access.csv')

    #print(file, "open access : OK")
#print the last
    df2
```
